```python
from bs4 import BeautifulSoup
import requests
import tushare as ts # pip install tushare
import pandas as pd
import numpy as np
import time
import sys
import io
import json
```

```python
stk_list = ts.get_stock_basics().name # get_china_a_share_stock_list
```

```python
stk_name = pd.DataFrame(stk_list).reset_index()
stk_name.head()
```

```python
for i in range(501):
    print(stk_name['name'][i])
```

```python
def get_news_titles(index, keyword):
    with io.open('No.{}_baidu.txt'.format(index),'w',encoding='utf-8') as g:
        keyword = keyword
        page_id = 1
        ids = []
        while True:
            url = 'http://news.baidu.com/ns?word=title%3A%28{}%29&pn={}&cl=2&ct=0&tn=newstitledy&rn=50&ie=utf-8&bt=1514736000&et=1531839599'.format(keyword,0+50*(page_id-1))
            r = requests.get(url)
            html = r.text
            soup = BeautifulSoup(html, 'html.parser') 
            div_items = soup.find_all('div', class_='result') 
            for i in div_items:
                # get titles
                title = i.contents[1].find('a').contents
                title = [x.string for x in title]
                title = ''.join(title).strip()
                # get publish time
                time = i.contents[3].contents[0].string.strip().split('\t')[-1]
                id_pos = str(i).find('id')
                # get news id
                ID = str(i)[id_pos+4:id_pos+7].replace('"','').replace('>','')
                output = str({'keyword':keyword,'title':title,'time':time})
                g.write(output)
                g.write('\n')
                if ID in ids:
                    print('finished!!!')
                    return
                ids.append(ID)
                print('id: {}'.format(ID))
                print('title: {}'.format(title))
                print('time: {}'.format(time))
            page_id += 1
            print('now on page: {}'.format(page_id))
```

```python
get_news_titles(500, stk_name['name'][500])
```

```python
for i in range(436, 500):
    get_news_titles(i, stk_name['name'][i])
```

```python
df = pd.DataFrame(columns=['keyword','title','time'])
for j in range(501):
    with open('No.{}_baidu.txt'.format(j)) as f:
        lines = f.readlines()
    for i in range(len(lines)):
        diction = eval(lines[i])
        d = pd.DataFrame(diction, index=[i])
        #print(d)
        df = pd.concat([df, d])   
```

```python
merged = pd.merge(stk_name, df, left_on='name', right_on='keyword', how = 'right')
merged.drop(['keyword'], axis=1, inplace=True)
```

```python
merged.to_csv('BAIDU_NEWS.txt', header=None, index=None, sep=' ', mode='a')
```
