```python
stk_list = [x for x in stk_list if x not in data.sheet_names]
```

```python
writer = pd.ExcelWriter('./stock_price.xlsx', engine = 'xlsxwriter')
```

```python
a = time.time()
j = 0
for i in stk_list:
    price = ts.get_k_data(i,start='2015-01-01') # save the price dataframe of a specific stock into the excel file
    price.to_excel(writer, sheet_name = i)
    if j % 20 == 0:
        print('{}:{} finished - {}'.format(j,len(stk_list),i))
        b = time.time()
        print('{}s taken, about {}s left for completion'.format((b-a),(len(stk_list)-j)*((b-a)/20)))
        a = time.time()
    j += 1
```
